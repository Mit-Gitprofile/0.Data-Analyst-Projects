```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
store = pd.read_csv("F:\Github_Desktop\Data_Analysis\Exam\supermarket_sales.csv")
print(store.head(10).to_string())
print(store.columns)
```

```python

plot = sns.barplot(data=store,x="City",y="Total",color="#7AE2CF")

plot.set_ylabel("Total Sales")
plt.title("City wise sales ...")
```

```python

city_count=store['City'].value_counts()

plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%')
# plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%',explode=[0.1,0,0.1])
plt.title("Transaction Distribution by City")
```

```python
sns.displot(data=store, x='Payment', kind='hist',element='step',col='Gender' , row="Customer type" ,color="#ba91ff")
```

```python

store_top_50=store.head(50)
sns.scatterplot(x='Quantity',data=store_top_50,y='Unit price', hue=store_top_50['City'])
plt.title('Unit Price vs Quantity')
plt.legend(loc="upper left")
```

```python
Product_count=store['Product line'].value_counts()

plt.pie(Product_count, labels=Product_count.index, autopct='%1.1f%%')
plt.title("Category Wise Quentity Sold")
```

```python
sns.boxplot(x='Gender',y='Quantity',data=store ,color="#E14434")
sns.boxplot(x='Customer type',y='Quantity',data=store ,color="#FEFBC7")
```

```python
branch_count=store['Branch'].value_counts()
branch_count
```

```python

plt.bar(branch_count.index, branch_count, color="#FF3F33")

plt.title("sales per Branch")
plt.xlabel("Branch")
plt.ylabel("sales")

```

```python

daily_sales = store.groupby('Date')['Total'].sum()


plt.plot(daily_sales.index, daily_sales, marker='o', color='blue')
plt.title("Daily Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid()
plt.xticks(rotation=55)
```

```python
avg_rating = store.groupby('Date')['Rating'].mean()
plt.plot(avg_rating.index, avg_rating, marker='.', color='green')
plt.title("Daily Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid()
plt.xticks(rotation=55)
```
